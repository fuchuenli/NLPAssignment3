```python
# Import  libray
import pandas as pd
import json
```

```python
# Load the metadata
with open("../input/CORD-19-research-challenge/metadata.csv", "r") as raw_df:
    whole_df = pd.read_csv(raw_df)
```

```python
# Exclude any subjects without corpus provided
whole_df_clean = whole_df.dropna(subset=["pdf_json_files", "pmc_json_files"], how="all")
whole_df_clean
```

```python
# Random sampling 6300 subject
part_df = whole_df_clean.sample(10000, random_state=0)
part_df
```

```python
# Read the text in the json file and extract the body text
def get_text(row):
    all_text = ""
    if pd.isna(row["pmc_json_files"]) == False:
        code = str(row["pmc_json_files"])
        
    else:
        code = str(row["pdf_json_files"])
    
    if ";" in code:
        code = code.split(";")[0]
        
    with open("../input/CORD-19-research-challenge/" + code, "r") as article:
            body_text = json.load(article)["body_text"]
            for item in body_text:
                all_text = all_text + "\n\n" + item["text"]
    return all_text.strip()
```

```python
# Add the corpus body text into the dataframe
part_df["corpus"] = part_df.apply(get_text, axis=1)
part_df["corpus"]
```

```python
# Showing the sampled modified dataframe
part_df
```

```python
# Extract the subset of the dataframe
select_df = part_df.loc[:, ["cord_uid", "title", "abstract", "authors", "publish_time", "corpus"]]
```

```python
# Convert the dataframe to csv and export it
select_df.to_csv("Corpus_10k.csv", index=False)
```
